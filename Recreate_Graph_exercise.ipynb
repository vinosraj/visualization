#Question - 

##How do cereal rating compare between different manufacturers. And is there are difference between hot or cold cereals.?

```python
fig,axes = plt.subplots(nrows=2, figsize=(14,8))
plt.xticks(rotation=90)
sns.barplot(data=df, x= 'Manufacturer', y ='Rating of cereal', ax=axes[0])
sns.barplot(data=df, x= 'type', y ='Rating of cereal', ax=axes[1])

fig.suptitle('Comparing Cereal Rating by Type & Manufacturer',y= 1.05)
plt.tight_layout()
```

#Does Sugar content or dietary fiber content impact the rating of the cereal?



```python
#grams of sugars
#grams of dietary fiber                          
```

```python
sns.set(rc={'figure.figsize':(10,8)})

ax=sns.scatterplot(data=df, y='grams of sugars', x='Rating of cereal', hue='Manufacturer')
ax.legend(bbox_to_anchor=[1,1])
```

Interesting the rating of the goes down with the increase of grams of sugar.

```python
sns.set(rc={'figure.figsize':(10,8)})

ax=sns.scatterplot(data=df, y='grams of dietary fiber', x='Rating of cereal', hue='Manufacturer')
ax.legend(bbox_to_anchor=[1,1])
```

Although there seems to be a slight increase of rating with the increase of fiber content, this wouldnt be conclusive.

```python
#corr = df.corr()
sns.heatmap(df.corr(), annot=True);
```

There seems to be a moderate correlation based on the heat map between Dietary fibers and cereal rating.
